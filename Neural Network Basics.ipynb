# Task - 2

* Use User defined inputs, weights, and bias & write a code for single perceptron and apply activation function.



```python
import numpy as np

# Defining the sigmoid activation function, it decides the output after summation is done
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def step_function(x):
    return 1 if x >= 0 else 0

# Defining the summation function, it adds the dot product of input, weights to bias
def perceptron(inputs, weights, bias):
    # Calculate the total sum of inputs
    total_sum = np.dot(inputs, weights) + bias
    
    # Applying the sigmoid activation function
    output = sigmoid(total_sum)
    
    return output

# Taking user-defined inputs
inpNum = int(input("Number of inputs to Neuron: "))
tempInp = []
for i in range(inpNum):
    i = float(input("Enter Input Number " + str(i + 1) + ": "))
    tempInp.append(i)
inputs = np.array(tempInp)
print("The inputs are:", inputs)

# Taking user-defined weights
tempWeight = []
for i in range(inpNum):
    i = float(input("Enter Weight for input " + str(inputs[i]) + ": "))
    tempWeight.append(i)
weights = np.array(tempWeight)
print("The weights are:", weights)

# Taking user-defined bias value
bias = float(input("Enter Bias Value:"))
print("The bias value is:", bias)

# Calculating the output of the perceptron fired
output = perceptron(inputs, weights, bias)
sigout = sigmoid(output)
print("Output of sigmoid function is: ", sigout)
result = step_function(output)
print("Output after computation is:", result)
```

# Task - 4 (a) (Separate Layer Method)
* Multi-layer Perceptron with two layers. Each layer having 3 perceptrons with user-defined values of inputs, weights & bias.



```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def summ_fn(sinput, weights, biases):
    summation = np.dot(sinput, weights) + biases
    return sigmoid(summation)

# Taking user-defined inputs
tempInp = []
for i in range(3):
    i = float(input("Enter Input Number " + str(i + 1) + ": "))
    tempInp.append(i)
inputs = np.array(tempInp)

# User Defined Weights and Bias Values for Layer 1
w1 = np.empty((3,3))
for i in range(3):
    for j in range(3):
        weight_input = float(input(f"Enter the value for w1[{i}][{j}]: "))
        w1[i][j] = weight_input

b1 = float(input("Enter Bias Value for Layer 1:"))

# User Defined Weights and Bias Values for Layer 2
w2 = np.empty((3,3))
for i in range(3):
    for j in range(3):
        weight_input = float(input(f"Enter the value for w2[{i}][{j}]: "))
        w2[i][j] = weight_input

b2 = float(input("Enter Bias Value for Layer 2:"))

# Forward propagation for Layer 1
outL1 = summ_fn(inputs, w1, b1)

# Forward propagation for Layer 2
outL2 = summ_fn(outL1, w2, b2)

print("---- Following are the resutls ----")
print("\nInput values for Layer 1 Perceptrons:")
print(inputs)
print("\nWeight values for Layer 1 Perceptrons:")
print(w1)
print("\nBias values for Layer 1 Perceptrons:")
print(b1)
print("\nOutput values for Layer 1 Perceptrons:")
print(outL1)

print("\nInput values for Layer 2 Perceptrons:")
print(outL1)
print("\nWeight values for Layer 2 Perceptrons:")
print(w2)
print("\nBias values for Layer 2 Perceptrons:")
print(b2)
print("\nOutput values for Layer 2 Perceptrons:")
print(outL2)
```

# Task - 6 (a) (Separate Layer Method)
* Multi-layer Perceptron with two layers. Each layer having VARIABLE perceptrons with random values of inputs, weights & bias.

# Task - 7
* Real-life binary problem with dataset as an input having 3 features with 4 data-points. The weight & bias values are generated randomly for a single layer, single perceptron.



```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def threshold(x):
    return 1 if x>0.5 else 0

# These are the 4 datapoints in the dataset
datapoints = np.array([
    [1,0,0],
    [1,0,1],
    [0,1,0],
    [1,1,1]
])

# Displaying inputs
print("The inputs are:\n", datapoints)

weights = np.random.rand(3)
print("\nThe random weights are:\n", weights)
bias = np.random.rand(1)
print("\nThe bias value is: ", bias)

# These are the target labels for the above datapoints respectively
tLabels = np.array([0,0,0,1])
print("\nThe target class/lables are: ", tLabels)

# Assigning weights for each input in the datapoint & calculating the dot product
for i in range(4):
    z = np.dot(datapoints[i], weights) + bias
    output = sigmoid(z)
    output = threshold(output)
    print(f"\nThe prediction for sample {i + 1} is:", output)
```

# Task - 8
* Real-life binary problem with dataset as an input having 3 features with 4 data-points. The weight & bias values are generated randomly for a multilayer layer, multi perceptron.



```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def threshold(x):
    return 1 if x>0.5 else 0
def summ_fn(sinput, weights, bias):
    summation = np.dot(sinput, weights) + bias
    return sigmoid(summation)

# These are the 4 datapoints in the dataset
datapoints = np.array([
    [1,0,0],
    [1,0,1],
    [0,1,0],
    [1,1,1]
])

# Displaying inputs
print("The inputs are:\n", datapoints)

w1 = np.random.rand(3,3)
print("\nThe weight values for Layer 1 is:\n", w1)
w2 = np.random.rand(3,3)
print("\nThe weight values for Layer 2 is:\n", w2)
b1 = np.random.rand(1)
print("\nThe bias value for Layer 1 is: ", b1)
b2 = np.random.rand(1)
print("\nThe bias value for Layer 2 is: ", b2)
wout = np.random.rand(3)
print("\nThe weight values for Output Layer is:\n", wout)
bout = np.random.rand(1)
print("\nThe bias value for Output Layer is: ", bout)
# These are the target labels for the above datapoints respectively
tLabels = np.array([0,0,0,1])
print("\nThe target class/lables are: ", tLabels)

for i in range(4):
    out1 = summ_fn(datapoints[i], w1, b1)
    out2 = summ_fn(out1, w2, b2)
    final = threshold(summ_fn(out2, wout, bout))
    print(f"\nOutput of layer 1 for sample {i + 1} is:", out1)
    print(f"Output of layer 2 for sample {i + 1} is:", out2)
    print(f"Prediction for sample {i + 1} is:", final)


```
